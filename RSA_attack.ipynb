```python
def decrypt(encrypted_message, e):
  num = round(pow(encrypted_message, 1/e))
  return str(bytes([x for x in num.to_bytes(1000, byteorder='little') if x != 0]), encoding='utf-8')

m = decrypt(109599938775622399587345269212768768, 3)
print("The encrypted message is:\n" + m)
```

```python
def find_continued_fraction(N, e):
  import fractions
  arr = []
  r = fractions.Fraction(e, N)
  count = 0
  while r.denominator != 0:
    count += 1
    i = r.numerator // r.denominator
    arr.append(i)
    f = r - i
    if f == 0:
      break
    r = fractions.Fraction(1, f)
  return arr

def find_convergents(arr):
  h = []
  k = []
  # will not include negatives indices in the returned arrays
  for i in range(len(arr)):
    if(i == 0):
      hn = arr[i]
      kn = 1
    elif(i == 1):
      hn = arr[i] * h[i-1] + 1
      kn = arr[i] * k[i-1]
    else:
      hn = arr[i] * h[i-1] + h[i-2]
      kn = arr[i] * k[i-1] + k[i-2]
    h.append(hn)
    k.append(kn)
  return h, k

def solve(a, b, c):
    from decimal import Decimal, getcontext
    getcontext().prec = 1000
    a, b, c = Decimal(a), Decimal(b), Decimal(c)
    return (int((-b + (b**2 - 4*a*c).sqrt())/(2*a)), int((-b - (b**2 - 4*a*c).sqrt())/(2*a)))

def run_attack(N, e):
  continued_fraction = find_continued_fraction(N, e)
  h, k = find_convergents(continued_fraction)
  for i in range(len(continued_fraction)):
    a = h[i]
    b = k[i]
    if a == 0:
      continue
    potential_phi = (e * b - 1) // a
    p, q = solve(1, -(N- potential_phi +1), N)
    if p * q == N:
      return p,q
  # if no primes p, q are found
  return 0,0

p, q = run_attack(90581, 17993)
print("For K1 we have the primes " + str(p) + " and " + str(q))
```
