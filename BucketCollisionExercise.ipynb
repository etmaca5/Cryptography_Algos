```python
# problem 0b
import random
b = 10000
num_trials = 1000
n_vals = range(115, 126)
n_results = [0] * len(n_vals)

def simulate(n):
  buckets = [0] * b
  for i in range(n):
    idx = random.randint(0, b - 1)
    if buckets[idx] == 1:
      # 0 indicating a simulation with a collision
      return 0
    buckets[idx] = 1
  # 1 indicating a simulation with no collisions
  return 1

for i in range(num_trials):
  for j in range(len(n_vals)):
    n_results[j] += simulate(n_vals[j]) / num_trials
for j in range(len(n_vals)):
  print("n = " + str(n_vals[j]) + " empirical probability of no collision: " + str(n_results[j]))
```

```python
import matplotlib.pyplot as plt

def theoretical_probability(b, n):
    prob = 1
    for i in range(n):
        prob *= (b - i) / b
    return prob

theoretical_probs = []

for n in n_vals:
    theoretical_probs.append(theoretical_probability(b, n))

plt.figure(figsize=(10, 6))
plt.plot(n_vals, theoretical_probs, label='Theoretical Probability', marker='o')
plt.plot(n_vals, n_results, label='Empirical Probability', marker='x')
plt.xlabel('Number of Names (n)')
plt.ylabel('Probability of No Collision')
plt.title('Theoretical vs Empirical Probability of No Collisions in Hashing')
plt.legend()
plt.grid(True)
plt.show()
```

```python
n_vals = [i for i in range(261)]
theoretical_probs = []
empirical_probs = []

for n in n_vals:
    theoretical_probs.append(theoretical_probability(b, n))
    empirical_val = 0
    for i in range(num_trials):
      empirical_val += simulate(n) / num_trials
    empirical_probs.append(empirical_val)

plt.figure(figsize=(10, 6))
plt.plot(n_vals, theoretical_probs, label='Theoretical Probability')
plt.plot(n_vals, empirical_probs, label='Empirical Probability')
plt.xlabel('Number of Names (n)')
plt.ylabel('Probability of No Collision')
plt.title('Theoretical vs Empirical Probability of No Collisions in Hashing')
plt.legend()
plt.grid(True)
plt.show()
```
